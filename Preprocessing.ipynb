```python
import pandas as pd
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import string
import data_awal as da
import re # impor modul regular expression
```

```python
filepath_dict = {'data':   'data/data_tsv.txt'}
```

```python
df_list = []
for source, filepath in filepath_dict.items():
    df = pd.read_csv(filepath, names=['sentence'])
    df_list.append(df)

df = pd.concat(df_list)
```

```python
sentences = da.df['sentence']
output_file = 'data/gojekindonesiafix.csv'
n = 0
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
for isi in sentences:
    tanpa_angka = re.sub(r"\d+", "", isi)
    katabersih = re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tanpa_angka)
    katadasar = stemmer.stem(katabersih)
    
    tokens = word_tokenize(katadasar)
    listStopwords = set(stopwords.words('indonesian'))
    hasil = []
    s = " "
    
    for t in tokens:
        if t not in listStopwords:
            hasil.append(t)
    akhir = s.join(hasil)

    print(akhir)

    with open(output_file, 'a') as f:
        output_string = akhir + "\t" + "\n"
        f.write(output_string)

    n = n + 1
```
